check out this Pokemon API https://pokeapi.co/
Use the requests package to connect to this API and get and store data for 20 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny', attack base_stat, hp base_stat, defense base_stat

```python
def catch_em_all():
    pokemon_names = ['arceus', 'pikachu', 'palkia', 'blaziken', 'rayquaza', 'dragonite', 'snorlax', 
                     'emboar', 'entei', 'lucario',
                     'groudon', 'mewtwo', 'zapdos', 'solgaleo', 'charizard', 'dialga', 
                     'cobalion', 'zoroark', 'raikou', 'gyarados', 'daxter', 'sly cooper']

    pokemon_data = []
    for name in pokemon_names:
        response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{name}')
        if response.ok:
            data = response.json()
            name = data.get('name')
            abilities = [ability_data.get('ability').get('name') for ability_data in data.get('abilities')]
            base_experience = data.get('base_experience')
            front_shiny_sprite = data.get('sprites').get('front_shiny')
            base_stats = {}
            for stat in data.get('stats'):
                stat_name = stat.get('stat').get('name')
                if stat_name in ['hp', 'attack', 'defense']:
                    base_stat_value = stat.get('base_stat')
                    base_stats[stat_name] = base_stat_value
            
            pokemon_info = {
                'name': name,
                'front_shiny_sprite': front_shiny_sprite,
                'ability': abilities,
                'base_xp': base_experience,
                'base_stats': base_stats
            }
            pokemon_data.append(pokemon_info)
        else:
            print(f"The {name} escaped from the pokeball.")
    return pokemon_data

catch_em_all()

```
